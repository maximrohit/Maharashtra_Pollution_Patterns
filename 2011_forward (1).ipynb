```python
Base_2011=Base_PoP_ext[Base_PoP_ext.Year.isin([2011])]
```

```python
years=list(range(2012,2016))
```

```python
[2011,years]
```

```python
lookup=pd.DataFrame({"match":2011, "years":years})
```

```python
results=pd.merge(Base_2011, lookup, how='inner', left_on=['Year'], right_on=['match'],suffixes=('_ori', '_cal'))
```

```python
results.columns
```

```python
results['Male_curr']=(results.Male_curr+((results.years-results.Year)*results.Male_FDiff))
```

```python
results['Female_curr']=(results.Female_curr+((results.years-results.Year)*results.Female_FDiff))
results['Total_curr']=(results.Total_curr+((results.years-results.Year)*results.Total_FDiff))
results['Population_Density_curr']=(results.Population_Density_curr+((results.years-results.Year)*results.Population_Density_FDiff))
results['Sex_Ratio_curr']=(results.Sex_Ratio_curr+((results.years-results.Year)*results.Sex_Ratio_FDiff))
results['Literacy_Rate_curr']=(results.Literacy_Rate_curr+((results.years-results.Year)*results.Literacy_Rate_FDiff))
```

```python
results.head()
```

```python
Base_PoP_merge=results[['District_Name', 'years',  'Male_curr', 'Female_curr','Total_curr', 'Population_Density_curr', 'Sex_Ratio_curr','Literacy_Rate_curr']]
```

```python
Base_PoP_merge.columns=['District_Name', 'Year', 'Male', 'Female', 'Total','Population_Density', 'Sex_Ratio', 'Literacy_Rate']
```
