```python
#Improting the PCA module
from sklearn.decomposition import PCA
pca = PCA(svd_solver='randomized', random_state=42)
```

```python
#Doing the PCA on the train data
pca.fit(YearClusterDataNSo2)
```

```python
pca.components_
```

```python
colnames = list(YearClusterDataNSo2.columns)
pcs_df = pd.DataFrame({'PC1':pca.components_[0],'abs_PC1':abs(pca.components_[0]),'PC2':pca.components_[1],'abs_PC2':abs(pca.components_[1]),'PC3':pca.components_[2],'abs_PC3':abs(pca.components_[2]),'PC4':pca.components_[3],'abs_PC4':abs(pca.components_[3]), 'Feature':colnames})
pcs_df.head()
```

```python
pcs_df.sort_values(by='abs_PC1',ascending=False).head(30)
```

```python
pcs_df.sort_values(by='abs_PC2',ascending=False)
```

```python
%matplotlib inline
fig = plt.figure(figsize = (8,8))
plt.scatter(pcs_df.PC1, pcs_df.PC2)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
for i, txt in enumerate(pcs_df.Feature):
    plt.annotate(txt, (pcs_df.PC1[i],pcs_df.PC2[i]))
plt.tight_layout()
plt.show()
```

```python
%matplotlib inline
fig = plt.figure(figsize = (8,8))
plt.scatter(pcs_df.PC1, pcs_df.PC3)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 3')
for i, txt in enumerate(pcs_df.Feature):
    plt.annotate(txt, (pcs_df.PC1[i],pcs_df.PC3[i]))
plt.tight_layout()
plt.show()
```

```python
pcs_df.sort_values(by='abs_PC3',ascending=False)
```

```python
pcs_df.sort_values(by='abs_PC4',ascending=False)
```

```python
pca.explained_variance_ratio_
```

```python
#Making the screeplot - plotting the cumulative variance against the number of components
%matplotlib inline
fig = plt.figure(figsize = (12,8))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```
