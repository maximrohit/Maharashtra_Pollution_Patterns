```python
NO2COL=['City_Town_Village_Area','Type_of_Location','month_number','NO2_Median','Median_NO2_MOM_diff','NO2_P10','P10_NO2_MOM_diff','NO2_P90','P90_NO2_MOM_diff','NO2_Median_YOY_diff','NO2_P10_YOY_diff','NO2_P90_YOY_diff','count.thousands']

SO2COL=['City_Town_Village_Area','Type_of_Location','month_number','SO2_Median','Median_SO2_MOM_diff','SO2_P10','P10_SO2_MOM_diff','SO2_P90','P90_SO2_MOM_diff','SO2_Median_YOY_diff','SO2_P10_YOY_diff','SO2_P90_YOY_diff','count.thousands']

RSPMPM10COL=['City_Town_Village_Area','Type_of_Location','month_number','RSPM_PM10_Median','Median_RSPM_PM10_MOM_diff','RSPM_PM10_P10','P10_RSPM_PM10_MOM_diff','RSPM_PM10_P90','P90_RSPM_PM10_MOM_diff','RSPM_PM10_Median_YOY_diff','RSPM_PM10_P10_YOY_diff','RSPM_PM10_P90_YOY_diff','count.thousands']
```

```python
corrmat = Base_polpop_iell_carsales[NO2COL].corr() 
  
f, ax = plt.subplots(figsize =(16, 8)) 
sns.heatmap(corrmat, ax = ax, cmap ="YlGnBu", linewidths = 0.1)
```

```python
corrmat = Base_polpop_iell_carsales[NO2COL].corr() 
  
cg = sns.clustermap(corrmat, cmap ="YlGnBu", linewidths = 0.1); 
plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(), rotation = 0) 
  
cg 
```
