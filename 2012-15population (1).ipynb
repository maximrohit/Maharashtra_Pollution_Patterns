```python
Base_PoP_ext = pd.read_csv("C:\MS\EDA\Base_PoP_merge201215.csv")
```

```python
Base_Poll_Popu= pd.read_csv("C:\MS\EDA\Base_Poll_Popu_calculate_pop2011onwards.csv")
```

```python
Base_Poll_Popu['year_number']=Base_Poll_Popu.year_number.map(int)
```

```python
Base_Poll_Popu_2012=Base_Poll_Popu.merge(Base_PoP_ext,how='left', suffixes=('_ori', '_cal'),left_on=['City_Town_Village_Area','year_number'], right_on=['District_Name','Year'])
```

```python
Base_Poll_Popu_2012.head(n=2)
```

```python
Base_Poll_Popu_2012.loc[Base_Poll_Popu_2012.year_number==2012,['Male_ori','Male_cal']].head(n=2)
```

```python
Base_Poll_Popu_2012.loc[(Base_Poll_Popu_2012.Male_ori.map(np.isnan)) & (Base_Poll_Popu_2012.year_number>=2012)].head(n=2)
```

```python
Base_Poll_Popu_2012.loc[(Base_Poll_Popu_2012.Male_ori.map(np.isnan)) & (Base_Poll_Popu_2012.year_number>=2012),'Male_ori']=Base_Poll_Popu_2012.loc[(Base_Poll_Popu_2012.Male_ori.map(np.isnan)) & (Base_Poll_Popu_2012.year_number>=2012),'Male_cal']
```

```python
Base_Poll_Popu_2012.loc[(Base_Poll_Popu_2012.Year_ori.map(np.isnan)) & (Base_Poll_Popu_2012.year_number>=2012),'Year_ori']=Base_Poll_Popu_2012.loc[(Base_Poll_Popu_2012.Year_ori.map(np.isnan)) & (Base_Poll_Popu_2012.year_number>=2012),'Year_cal']
Base_Poll_Popu_2012.loc[(Base_Poll_Popu_2012.Female_ori.map(np.isnan)) & (Base_Poll_Popu_2012.year_number>=2012),'Female_ori']=Base_Poll_Popu_2012.loc[(Base_Poll_Popu_2012.Female_ori.map(np.isnan)) & (Base_Poll_Popu_2012.year_number>=2012),'Female_cal']
Base_Poll_Popu_2012.loc[(Base_Poll_Popu_2012.Total_ori.map(np.isnan)) & (Base_Poll_Popu_2012.year_number>=2012),'Total_ori']=Base_Poll_Popu_2012.loc[(Base_Poll_Popu_2012.Total_ori.map(np.isnan)) & (Base_Poll_Popu_2012.year_number>=2012),'Total_cal']
Base_Poll_Popu_2012.loc[(Base_Poll_Popu_2012.Population_Density_ori.map(np.isnan)) & (Base_Poll_Popu_2012.year_number>=2012),'Population_Density_ori']=Base_Poll_Popu_2012.loc[(Base_Poll_Popu_2012.Population_Density_ori.map(np.isnan)) & (Base_Poll_Popu_2012.year_number>=2012),'Population_Density_cal']
Base_Poll_Popu_2012.loc[(Base_Poll_Popu_2012.Sex_Ratio_ori.map(np.isnan)) & (Base_Poll_Popu_2012.year_number>=2012),'Sex_Ratio_ori']=Base_Poll_Popu_2012.loc[(Base_Poll_Popu_2012.Sex_Ratio_ori.map(np.isnan)) & (Base_Poll_Popu_2012.year_number>=2012),'Sex_Ratio_cal']
```
